# 3. Image example visualization

```python
# function that reads images using cv2
def read_img(file_path):
    img_arr = cv2.imread(file_path)
    return cv2.cvtColor(img_arr, cv2.COLOR_BGR2RGB) # turn BGR into RGB with cvtColor
```

```python
# visualize 3 images randomly in each image folder
# ready image array
img_arrs = []
img_num = range(0,300)

for i in random.sample(img_num,3):
    
    img_arrs.append(read_img(snack0[i]))
    img_arrs.append(read_img(snack1[i]))
    img_arrs.append(read_img(snack2[i]))
```

```python
rows = 3
columns = 3

fig, axes = plt.subplots(nrows=rows, ncols=columns, figsize=(columns*3, rows*3))

# output images
for num in range(1, rows*columns+1):
    
    fig.add_subplot(rows, columns, num) 
    idx = num - 1 # index
    
    plt.imshow(img_arrs[idx], aspect='auto')
    plt.xlabel(f'{img_arrs[idx].shape}', fontsize=12)

fig.tight_layout() # adjust the spacing between images

# title
cols = ['snack0', 'snack1', 'snack2']

for folder_idx, ax in enumerate(axes[0]):
    ax.set_title(cols[folder_idx])

# remove xtick, ytick
for idx, ax in enumerate(axes.flat):
    ax.set_xticks([])
    ax.set_yticks([])
```

```python
# count test dataset 
snack0_test_count = round(len(snack0)*0.2)
snack1_test_count = round(len(snack1)*0.2)
snack2_test_count = round(len(snack2)*0.2)
snack3_test_count = round(len(snack3)*0.2)
snack4_test_count = round(len(snack4)*0.2)
snack5_test_count = round(len(snack5)*0.2)
snack6_test_count = round(len(snack6)*0.2)
snack7_test_count = round(len(snack7)*0.2)
snack8_test_count = round(len(snack8)*0.2)
snack9_test_count = round(len(snack9)*0.2)
```

```python
def split( img_list, test_count, train_path, test_path):
  
    test_files=[]
    for i in random.sample( img_list, test_count ):
        test_files.append(i)

    # create train/test list
    train_files = [x for x in img_list if x not in test_files]

    for k in train_files:
        try:
            shutil.copy(k, train_path)
        except:
            pass

    for c in test_files:
        try:
            shutil.copy(c, test_path)
        except:
            pass
```

```python
# path settings
snack0_train_path='./train/꼬깔콘'
snack0_test_path='./test/꼬깔콘'

snack1_train_path='./train/꼬북칩'
snack1_test_path='./test/꼬북칩'

snack2_train_path='./train/맛동산'
snack2_test_path='./test/맛동산'

snack3_train_path='./train/새우깡'
snack3_test_path='./test/새우깡'

snack4_train_path='./train/에이스'
snack4_test_path='./test/에이스'

snack5_train_path='./train/오징어 땅콩'
snack5_test_path='./test/오징어 땅콩'

snack6_train_path='./train/카스타드'
snack6_test_path='./test/카스타드'

snack7_train_path='./train/포카칩'
snack7_test_path='./test/포카칩'

snack8_train_path='./train/프링글스'
snack8_test_path='./test/프링글스'

snack9_train_path='./train/홈런볼'
snack9_test_path='./test/홈런볼'

# split 
split(snack0, snack0_test_count, snack0_train_path, snack0_test_path)
split(snack1, snack1_test_count, snack1_train_path, snack1_test_path)
split(snack2, snack2_test_count, snack2_train_path, snack2_test_path)
split(snack3, snack3_test_count, snack3_train_path, snack3_test_path)
split(snack4, snack4_test_count, snack4_train_path, snack4_test_path)
split(snack5, snack5_test_count, snack5_train_path, snack5_test_path)
split(snack6, snack6_test_count, snack6_train_path, snack6_test_path)
split(snack7, snack7_test_count, snack7_train_path, snack7_test_path)
split(snack8, snack8_test_count, snack8_train_path, snack8_test_path)
split(snack9, snack9_test_count, snack9_train_path, snack9_test_path)
```

# 5. Data augmentation

```python
# path settings
snack_list = ['꼬깔콘', '꼬북칩', '맛동산', '새우깡', '에이스', '오징어 땅콩', '카스타드', '포카칩', '프링글스', '홈런볼']
file_path = '/content/drive/MyDrive/data/train'
save_path = '/content/drive/MyDrive/data/train'
    
random_augment = random.randrange(1, 6)

# data augmentation
for snack in snack_list:
    file_list = os.listdir(file_path + '/' + snack)
    augment_cnt = 1
    for f in file_list:
        
        try:
            random_augment = random.randrange(1, 5)
            image = Image.open(file_path + '/' + snack + '/' + f)
            
            # left right inversion
            if(random_augment == 1) :
                inverted_image = image.transpose(Image.FLIP_LEFT_RIGHT)
                inverted_image.save(save_path + '/' +  snack + '/' + snack + "inverted_"+str(augment_cnt) + ".jpg")

            # rotation
            elif (random_augment == 2) :
                rotated_image = image.rotate(random.randrange(-20, 20))
                rotated_image.save(save_path + '/' +  snack + '/' + snack + "rotated_" + str(augment_cnt) + ".jpg")

            # top bottom inversion
            elif(random_augment == 3) :
                inverted_top_image = image.transpose(Image.FLIP_TOP_BOTTOM)
                inverted_top_image.save(save_path + '/' +  snack + '/' + snack + "inverted_top_" + str(augment_cnt) + ".jpg")

            # black and white 
            elif(random_augment == 4) :
                gray_image = image.convert('L')
                gray_image.save(save_path + '/' +  snack + '/' + snack + "gray_" + str(augment_cnt) + ".jpg")

            # brightness
            elif(random_augment == 5) :
                brightness_image = ImageEnhance.Brightness(image).enhance(0.4)
                brightness_image.save(save_path + '/' +  snack + '/' + snack + "brightness_" + str(augment_cnt) + ".jpg")
            
            augment_cnt += 1
            
        except Exception as e:
            pass
```
